# Building up data summaries across categorical variables

```python
# 'pool' is a categorical variable. Count the number in each category
data['pool'].value_counts()
```

```python
# 'flo11geno' is another categorical variable. Let's do the same for it.
data.flo11geno.value_counts()  # remember we can use attribute access on data frames to get columns
```

```python
# average the CM scores for each strain
cm = ['cmA','cmB','cmC']
mean_cm = data[cm].mean(axis=1)
```

```python
# another way to do this would be to use the select method of dataframes
# this would allow us to select columns that match a certain pattern, without having to
# explicitly right out all the column names (useful especially with large data sets)

mean_cm_alt = data.select(lambda x: 'cm' in x, axis=1).mean(axis=1)
```

```python
sum(mean_cm != mean_cm_alt)  # make sure these gave us the same results
```

```python
# now add a new column to our data frame
data['mean_cm'] = mean_cm
data.head()  # note the new columns
```

```python
hist(data.mean_cm, color='maroon', alpha=0.5)
ylim(0,40)
```

```python
# get the mean and stddev of the colony morphology means for the complex strains
data.mean_cm[data.pool == 'C'].mean(), data.mean_cm[data.pool == 'C'].std()
```

```python
# Let's find the adhesion means

adhes = ['adhesA', 'adhesB', 'adhesC']
mean_adhes = data[adhes].mean(axis=1)
data['mean_adhes'] = mean_adhes
```
